读取垃圾的内容

```python
examples=[]
file_path =r'.\data_spam.csv'
with open(file_path,encoding='latin-1') as f:
    for line in f.readlines()[1:]:
        line = line.replace('"',' ').replace('\n','').strip()
        pos = line.find(',') 
        label, text = line[:pos].strip(),line[pos+1:]
        examples.append([text,label])
    f.close()
print(examples)
```

```python
from collections import Counter 
#X: instances, Y: labels
X,Y = list(zip(*examples))

label_counts = Counter(Y)
print(label_counts)
print(X[:10])
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer,WordNetLemmatizer
import string

import nltk
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
# Initialize stemmer and lemmatizer
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

# Define function to preprocess text
def preprocess_text(text):
    # Convert to lowercase
    text = text.lower()
    # 去除标点符号
    text = "".join([char for char in text if char not in string.punctuation])
   
    # 分词
    tokens = nltk.word_tokenize(text)
    # 词根还原：
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]
    stemmed_tokens = [stemmer.stem(token) for token in lemmatized_tokens]
    # Join tokens back into text string
    preprocessed_text = " ".join(stemmed_tokens)
    return preprocessed_text
```

```python
#经过去除标点符号、词根还原
X = [preprocess_text(text).split() for text in X]
print(X[:10])
```

```python
import itertools
from collections import Counter
#统计所有词出现的次数
word_count =Counter(list(itertools.chain(*X)))
#按次数从高到低排序
word_count = sorted(word_count.items(), key = lambda kv:(kv[1], kv[0]), reverse = True)
#输出总词数
print(len(word_count),word_count[:10])
```

```python
#对每个样本统计词频
word2id = {w:k for k,(w,_) in enumerate(word_count)}
id2word ={k:w for (w,k) in word2id.items()}
print(word2id)

cat2id ={category: k for k,(category,_) in enumerate(label_counts.items())}
id2cat={v:k for (k,v) in cat2id.items() }
print(cat2id,id2cat)
```

```python
#对每个样本进行处理
docs=[ list(Counter(x).items()) for x in X]
print(docs[:10])
```

```python
#构建Naive Bayes
import numpy as np
_alpha, _beta = 0.01, 0.001
#平滑，每个label增加了alpha
state_matrix = np.array(list(label_counts.values()))+ _alpha
#归一化
state_matrix = state_matrix/np.sum(state_matrix)
print(label_counts,state_matrix)

emit_matrix =np.full((len(label_counts), len(word2id)),_beta)

#统计各个标签下词出现的次数
for x, y in zip(docs,Y):
    for (w, c) in x:
        emit_matrix[cat2id[y]][word2id[w]] +=c

#归一化生成概率
emit_matrix = emit_matrix/np.sum( emit_matrix,axis=-1,keepdims=True)
print(emit_matrix, np.sum(emit_matrix,axis=-1))
```

```python
#Bayes prediction 
#[2* sample_size]
scores =np.array( [[  np.log(state_matrix[cat2id[y] ]) + np.sum([ c*np.log(emit_matrix[cat2id[y]][word2id[w]])
                                                                 for (w,c) in x ]) for x in docs ]
                                                                for y in cat2id.keys()])
#[sample_size,2]
scores = scores.transpose(1,0)
print(scores.shape)

#选择分数最大的label
predictions = np.argmax(scores, axis=-1)
print(predictions.shape)

target = np.array([ cat2id[y] for y in Y])
print(target.shape)
```

```python
#比较预测性能
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

print(classification_report(target, predictions))
SVCAcc = accuracy_score(predictions,target)
print('Naive Bayes accuracy is: {:.2f}%'.format(SVCAcc*100))
```
