#### 爬虫程序

##### 导包

```python
# -*- coding: utf-8 -*-
from bs4 import BeautifulSoup as bs  # 网页解析，获取数据
import re  # 正则表达式，进行文字匹配`
import urllib.request, urllib.error  # 制定URL，获取网页数据
```

##### 获取网页网址

```python
def get_soup(url):
    head = {
        'User_Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/124.0.0.0 Safari/537.36 Edg/124.0.0.0'
    }
    try:
        req = urllib.request.Request(url , headers=head) # 爬取网页
        resp = urllib.request.urlopen(req)
        html = resp.read().decode("utf-8") # 获取网页信息
        soup = bs(html , 'html.parser')
        soup = soup.find_all('div')
        return soup
    except:
        return "####"

def get_url():
    href = []
    url = "https://www.cauc.edu.cn/"
    head = {
        'User_Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/124.0.0.0 Safari/537.36 Edg/124.0.0.0'
    }

    # 获取li标签下的网址
    def get_li(url):
        temp = []
        try:
            res = urllib.request.Request(url , headers=head)
            resp = urllib.request.urlopen(res)
            html = resp.read().decode('utf-8')
            soup = bs(html , "lxml")
            urls_li = soup.select("a")
            for i in urls_li:
                s = str(i)
                if "htt" in s and "target" in s:
                    idx = s.index("htt")
                    idx1 = s.index("target")
                    s = s[idx:idx1 - 2]
                    print(s)
                    temp.append(s)
        finally:
                return "####" if not len(temp) else temp
    # url = "https://www.cauc.edu.cn/"找到所有的url
    href = get_li(url)
    href1 = href.copy()
    try:
        for i in href:
            t = get_li(i) # 只再深搜一次，多次非常慢
            if t == '####':
                continue
            for j in t:
                href1.append(j)
    finally:
        with open('total_url.txt' , 'w') as f: # 保存url文件
            for i in href1:
                f.write(i + "\n")
            f.close()
        return href1
```

##### 整理url

```python
def sort_url():
    with open('total_url.txt' , 'r') as f:
        for i in f.readlines():
            if 'cauc' in i and 'click' not in i:
                with open('check_url.txt' , 'a') as f1:
                    f1.write(i)
                    f1.close()
```

##### 获取中文字符

```python
def get_chinese():
    with open('check_url.txt' , 'r') as f:
        for i in f.readlines():
            soup = get_soup(i)
            if soup == "####":
                continue
            # 匹配中文、中文标点、数字
            find_china = re.compile(r'[(\u4e00-\u9fa5)(0-9)(\u3002|\uff1f|\uff01|\uff0c|\u3001|\uff1b|\uff1a|\u201c|\u201d|\u2018|\u2019|\uff08|\uff09|\u300a|\u300b|\u3010|\u3011|\u007e)]' , re.S)
            cont = re.findall(find_china , str(soup))
            s = ''.join(cont)
            with open('total_context.txt' , 'a' , encoding='utf8') as f1:
                f1.write(s + "\n")
                f1.close()
```

##### 主程序

```python
get_url()
sort_url()
get_chinese()
```
